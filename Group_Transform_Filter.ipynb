```python
df = pd.DataFrame({'Name': ['Tom', 'Nick', 'Krish', 'Jack', 'Mike', 'James', 'Ryan', 'David'],
        'FName': ["Nicolas", "Pooran", "Pooran", "Pooran", "Nicolas","Nicolas","Nicolas","Hendry"],
        'Class': ["A", "A", "B", "B", "A","C","C", "D"],
        'Age': [20, 21, 19, 18, 22,9,5,1]})
  
df 
```

```python
df_agg = df.groupby("Class")[["Class", "Age"]].aggregate({"Age":["min", "max"], "Class":"count"})

df_agg.reset_index()

df_agg.columns = ["_".join(col) for col in df_agg.columns]

df_agg
```

# Transform and groupby

1. Get a tag/value like above average without filtering and joining the data
2. Get a tag/value like under each group/s which is min value

```python
df.groupby("Class")["Age"].mean()
```

```python
# Above_class_avg_tag --> Tag generated if name is having age above average of class-- tag as Ture
df["Above_class_avg_tag"] = df.groupby("Class")["Age"].transform(lambda x: x-x.mean()) > 0
df.sort_values('Class')
```

```python
# Above_class_avg_value --> Get value for vove case
df["Above_class_avg_value"] = df.groupby("Class")["Age"].transform(lambda x: x-x.mean())
df.sort_values('Class')
```

```python
# Min_by_FName_class_tag --> Get value for vove case
df["Min_by_FName_class_tag"] = df.groupby(["FName", "Class"])["Age"].transform(lambda x: x-x.min()) == 0
df.sort_values(['FName','Class'])
```

```python
# Min_by_FName_class_value --> Get value for vove case
df["Min_by_FName_class_value"] = df.groupby(["FName", "Class"])["Age"].transform(lambda x: x-x.min())
df.sort_values(['FName','Class'])
```

```python
# FILTER DATA THAT IS HAVING MORE THAN 10 IN AGE
```

```python
df.groupby("Name").filter(lambda x: x["Age"].max() > 10).reset_index(drop = True)
```

```python
# FILTER DATA THAT IS HAVING MORE THAN 0 IN Min_by_FName_class_value
```

```python
df.groupby("Name").filter(lambda x: x["Min_by_FName_class_value"].max() > 0).reset_index(drop = True)
```
